```python
report_card_dict = {"Student names" : ["sahil","tanish","vivan","siddhant","siddhath"],
                    "physics_grade" : [34,88,92,76,90],
                    "chemistry_grade" : [22,88,55,77,90],
                    "maths_grade": [33,77,98,77,94],
                    "biology_grade":[44,56,67,87,77]
                   }
                   
report_card_dict
```

```python
# setup
import pandas as pd
```

```python
report_card_df = pd.DataFrame(report_card_dict)
report_card_df
```

## insights

- what is the average chemistry grades? **73**
- what is the average biology grade? **72.8**
- what is the lowest math grade ? **67**
- what is the highest physics grade? **92**
- how many scores are present in the math coluumn? **5**


```python
# Make a total grade column

report_card_df["total_grade"] = report_card_df["physics_grade"] + report_card_df["maths_grade"] + report_card_df["chemistry_grade"] + report_card_df["biology_grade"]
report_card_df
```

```python
report_card_df["average_grade"] = report_card_df["total_grade"]/4
report_card_df
```

```python
report_card_df.describe()
```

```python
#Saving The report card insights data frame
report_card_insights_df = report_card_df.describe().transpose()
report_card_insights_df
```

```python
# Saving the Report card Insights as a csv file
report_card_insights_df.to_csv("Report Card insights.csv")
```

```python
# If total grade is greater or equal to 140 print make the staus vale as pass
report_card_df.loc[report_card_df["total_grade"]>=140,"Status"] ="PASS"

#If total grade is less than 140 make the status value as fail
report_card_df.loc[report_card_df["total_grade"]<140,"Status"] ="FAIL"
report_card_df
```

```python
# Making a grade column
report_card_df.loc[(report_card_df["average_grade"] >=90),"Grade"] = "A"
report_card_df.loc[(report_card_df["average_grade"] >=70) & (report_card_df["average_grade"]<90),"Grade"] = "B"
report_card_df.loc[(report_card_df["average_grade"] >=50) & (report_card_df["average_grade"]<70),"Grade"] = "c"
report_card_df.loc[(report_card_df["average_grade"] >=35) & (report_card_df["average_grade"]<50),"Grade"] = "D"
report_card_df.loc[(report_card_df["average_grade"] <35),"Grade"] = "F"
```
