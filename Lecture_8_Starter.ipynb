# Functions

```python
#setup 
import pandas as pd 
```

```python
# get scores is a function that accpets a report card dictionary and a division name 
#converts 
def get_scores(report_card_dict,division_name):
#convert dictionary ibto data frame 
    report_card_df = pd.DataFrame(report_card_dict)

# add a divsion coulumn (e.g A,B,C)
    report_card_df["division"] =  division_name
# add a column called total score (0 to 400)
    report_card_df["total_grade"] = report_card_df["physics_scores"] + report_card_df["math_scores"] + report_card_df["chemistry_scores"]+report_card_df["information_technology_scores"]
# add a column called average score (0 to 100)
    report_card_df["average_score"] = report_card_df["total_grade"]/4
# add a column called status (pass , fail)
    report_card_df.loc[report_card_df["total_grade"]>=140,"Status"] ="PASS"
    report_card_df.loc[report_card_df["total_grade"]<140,"Status"] ="FAIL"
# add a column called character grade (A , B , C , D , F)
    report_card_df.loc[(report_card_df["average_score"]>=90),"Grade"] = "A"
    report_card_df.loc[(report_card_df["average_score"]>=70) & (report_card_df["average_score"]<90),"Grade"] = "B"
    report_card_df.loc[(report_card_df["average_score"]>=50) & (report_card_df["average_score"]<70),"Grade"] = "C"
    report_card_df.loc[(report_card_df["average_score"]>=35) & (report_card_df["average_score"]<50),"Grade"] = "D"
    report_card_df.loc[(report_card_df["average_score"]<35),"Grade"] = "F"
# return the data frame 
    return report_card_df
```

## Get Email Function

```python
# Get emails 

def get_email(emails_dict,division_name):
    
   # converting dict into data frame
    emails_df=pd.DataFrame(emails_dict)
    
    # coverting a column called division
    emails_df["division"] = division_name
    
    # converting all the emails to lower case 
    emails_df["student_email"] =emails_df["student_email"].str.lower()
    
    #return the data frame 
    return emails_df
```

```python
div_a_email_df = get_email(div_a_emails_dictionary,"A")
div_a_email_df
```

## Get scores and email function

```python
def get_scores_and_emails_combined(scores_df,emails_df):
    report_card_with_emails_df = pd.merge(scores_df,emails_df, how = "left" , on = ["student_name","division"])
    return report_card_with_emails_df
```

## Get all division Score Sheet

```python
def get_all_division_score(div_a_df,div_b_df,div_c_df):
    all_div_df = pd.concat([div_a_df,div_b_df,div_c_df],ignore_index = True)
    return all_div_df 
```
